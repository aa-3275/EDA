#Analysis based on start time

```python
df.columns
```

```python
df.Start_Time
```

```python
df.Start_Time=pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time
```

```python
df.Start_Time.dt.hour
```

```python
counts_item=[]
for i in set(list(df.Start_Time.dt.hour)):
  counts_item.append(list(df.Start_Time.dt.hour).count(i))
counts_item
```

```python
Total_values=sum(counts_item)
```

```python
percentage_count=[]
for i in range (0, len(counts_item)):
  percentage_count.append(counts_item[i]/Total_values)

```

```python
percentage_count
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.scatter(list(set(list(df.Start_Time.dt.hour))), percentage_count, marker='o')
```

```python
plt.plot(list(set(list(df.Start_Time.dt.hour))), percentage_count)
```

```python
percentage_btw_3_to_5 = percentage_count[15]+ percentage_count[16] + percentage_count[17]
```
